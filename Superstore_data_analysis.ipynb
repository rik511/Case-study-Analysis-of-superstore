```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

```python
df.drop(columns='Postal Code', inplace=True)
df.head()
```
